### Expérimentations

<font size="+1" color="RED">**[Q]**</font> A partir de <code>data</code>, proposer et tester un classifieur permettant de prédire si le pays est de la région <code>European Union</code> ou non en utilisant la description composée uniquement des variables <code>Population (millions)</code>, <code>HDI</code>, <code>GDP/Capita</code>, <code>Total Ecological Footprint</code>, et <code>Earth Required</code>.

```python
data_set = ls.LabeledSet(5)
for index,ligne in data.iterrows():
    tmp = []
    tmp.append(ligne["Population (millions)"])
    tmp.append(ligne["HDI"])
    tmp.append(ligne["GDP/Capita"])
    tmp.append(ligne["Total Ecological Footprint"])
    tmp.append(ligne["Earths Required"])
    if(ligne["Region"] == "European Union"):
        data_set.addExample(tmp, 1)
    else:
        data_set.addExample(tmp, -1)
```

```python
def affiche_base(LS):
    """ LabeledSet
        affiche le contenu de LS
    """
    for i in range(0,LS.size()):
        print("Exemple "+str(i))
        print("\tdescription : ",LS.getX(i))
        print("\tlabel : ",LS.getY(i))
    return

# Affichage de la base
affiche_base(data_set)
```

```python
cp = cl.ClassifierPerceptron(5,0.01)

train, test = ut.split(data_set, 0.5)

cp.train(train)
print(cp.accuracy(train))
print(cp.accuracy(test))
```
